```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Load the dataset
df = pd.read_csv('C:/Users/User/Downloads/zoo.csv',header=0)

# Assuming the first column 'animal_name' should not be used in the feature set
X = df.drop(columns=['animal_name', 'type'])  # Dropping non-feature and target column
y = df['type']  # Target variable which is 'type'

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

# Initialize and fit the K-Neighbors Classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Predict on the test set
pred = knn.predict(X_test)

# Ensure pred and y_test are numpy arrays (especially if they've been modified or are not as expected)
pred = np.array(pred)
y_test = np.array(y_test)

# Output results and error analysis
print('Index\tPredicted\tActual')
for i in range(len(pred)):
    if pred[i] != y_test[i]:
        print(f"{i}\t{pred[i]}\t\t{y_test[i]} ****")

# Calculate and print accuracy and confusion matrix
accuracy = accuracy_score(y_test, pred)
conf_matrix = confusion_matrix(y_test, pred)
print(f'Accuracy Score: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
```
